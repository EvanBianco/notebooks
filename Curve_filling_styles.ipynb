## Display techniques for well logs

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
```

```python
% matplotlib inline
```

```python
x = np.random.normal(0, 1, 1100).cumsum()
x -= x.min()
x2 = 0.5 * x
y = np.arange(x.size)

fig = plt.figure(figsize=(10, 10), facecolor=None)
ax = fig.add_subplot(111)

# background image (grayscale static thing)
bgim = ax.imshow(np.random.rand(20, 20), cmap='Greys',
                 extent=[-200, 200, 0, 1500], interpolation='none', alpha = 0.25)

# fill-between image
im = ax.imshow(np.random.rand(y.max()/100, 1), extent=[x.min(), x.max(), y.min(), y.max()],
               cmap='gist_earth', aspect='auto', origin='lower')

# the right log (x) and the left log (x2)
paths = ax.fill_betweenx(y, x, x2, where= x2 <= x, facecolor='none')

# plot well as a green line
ax.plot([0, 0], [0, 1100], 'g-', lw=4, alpha=0.5)

patch = PathPatch(paths._paths[0], visible=False)

ax.add_artist(patch)

im.set_clip_path(patch)

ax.invert_yaxis()

plt.show()
```
