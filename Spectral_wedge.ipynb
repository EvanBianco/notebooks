```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
```

Set up the model. Warning — it gets very slow for numbers over 100. I tested everything with depth = length = width = 60.

```python
depth = 60. # z range
length = 60.# x range
width = 60. # y range
subwedges= 2
```

We will build the wedge layer by layer. Start with a blank slate. Rock = 1.

```python
value = 1.
m = np.ones((depth,length))
plt.imshow(m, interpolation='nearest')
plt.show()
```

Index into the array with *z* = depth, *x* = length, *v* = value (not needed). First a loop to generate the wedge 'contents', then another to put the top on. 

```python
value = 2.
for (z,x),v in np.ndenumerate(m):
    if z < (depth/3)*(1 + x/length):
        m[z,x] = value
        
value = 4.
for (z,x),v in np.ndenumerate(m):
    if z < depth/3:
        m[z,x] = value
     
plt.imshow(m, interpolation='nearest')
plt.show()
```

## Constant *x* slice.

Draw the bottom of the wedge.

```python
n = np.ones((depth,width))
value = 2.
for (z,y),v in np.ndenumerate(n):
    if z < 2*depth/3:
        n[z,y] = value

plt.imshow(n, interpolation='nearest')
plt.show()
```

```python
for i in range(subwedges):
    # Draw the bottom of the subwedge and flood with subwedge rock
    value = 3.
    for (z,y),v in np.ndenumerate(n):
        if z < (depth/3)*(2. - (2*i+1)/(2.*subwedges) + y/(2.*subwedges*width)):
            n[z,y] = value
    # Draw the top of the subwedge and flood with wedge rock
    value = 2.
    for (z,y),v in np.ndenumerate(n):
        if z < (depth/3)*(2. - (2*i+1)/(2.*subwedges) - y/(2.*subwedges*width)):
            n[z,y] = value
            
plt.imshow(n, interpolation='nearest')
plt.show()
```

Put the top in place:

```python
value = 4.
for (z,x),v in np.ndenumerate(n):
    if z < depth/3:
        n[z,x] = value

plt.imshow(n, interpolation='nearest')
plt.show()
```

## Full 3D array

Instead of looping over the entire array each time, we'll build a series of constant *x*-slices, as in the last section. 

```python
value = 1.
model = np.ones((depth,length,width))

for xslice in range(int(length)):# Iterate over the subwedges
    thickness = (depth/3)*(1 + xslice/length)
    
    # Draw the bottom of the wedge
    value = 2.
    for (z,y),v in np.ndenumerate(model[:,xslice,:]):
        if z < thickness:
            model[z,xslice,y] = value
        
    for i in range(subwedges):
        # Draw the bottom of the i-th subwedge and flood with subwedge rock
        value = 3.
        for (z,y),v in np.ndenumerate(model[:,xslice,:]):
            if z < (depth/3) + (thickness - depth/3) *(1 - (2*i+1)/(2.*subwedges) + y/(2.*subwedges*width)):
                model[z,xslice,y] = value
        # Draw the top of the i-th subwedge and flood with wedge rock
        value = 2.
        for (z,y),v in np.ndenumerate(model[:,xslice,:]):
            if z < (depth/3) + (thickness - depth/3)*(1 - (2*i+1)/(2.*subwedges) - y/(2.*subwedges*width)):
                model[z,xslice,y] = value

    # Put the top on
    value = 4.
    for (z,y),v in np.ndenumerate(model[:,xslice,:]):
        if z < depth/3:
            model[z,xslice,y] = value
```
